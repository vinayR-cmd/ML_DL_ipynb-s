```python
import spacy
nlp=spacy.load('en_core_web_sm')
```

```python
doc=nlp(u'I will google for the facebook')
doc.text
```

```python
doc[-1]
```

```python
doc[0].pos_
# coarse grained POS
```

```python
# fine grained POS
doc[-1].tag_
```

```python
# to get to know about the short forms 
spacy.explain('NN')
```

```python
for word in doc:
    print(word.text ,"---->", word.pos_, word.tag_, spacy.explain(word.tag_))
```

```python
doc4=nlp(u"i read books on nlp")
for word in doc4:
    print(word.text ,"---->", word.pos_, word.tag_, spacy.explain(word.tag_))
```

```python
doc5=nlp(u"i have read books on nlp")
for word in doc5:
    print(word.text ,"---->", word.pos_, word.tag_, spacy.explain(word.tag_))
```

```python
doc6=nlp(u"the quick brown fox jumped over the lazy dog")
from spacy import displacy
```
