```python
import pandas as pd

x=[1,2,3,4,5,6]
var=pd.Series(x)
print(var)
print(type(var))
print(var[2])

#  to change the index numbers according to ourselves
var=pd.Series(x,index=['a','b','c','d','e','f'])
print(var)

# to change the data ttype of the eleemnts of the series
var=pd.Series(x,index=['a','b','c','d','e','f'],dtype=float)
print(var)

var=pd.Series(x,index=['a','b','c','d','e','f'],dtype=float,name="python")
print(var)
```

```python
# converting the dictionary to the series data structure
dict={ "name": ['python','c','java','cpp'], "por":[30,40,50,60],"rank" :[1,2,3,4]}
var=pd.Series(dict)
print(var)
# due to the mix up of the differeent data types the data type os this serrie is objecct


# to make series with a single element
x=pd.Series(10,index=[1,2,3,4,5,6])
print(x)

# to add the series having the differnt number of elements
x=pd.Series([2,2,2,2,2,2,2,2,2,2])
y=pd.Series([2,2,2,2,2,2,2])
print(x+y)
# pandas alsop works with the missing data values
```

```python
# Daraframe => 2D data structure
import pandas as pd
x=[1,2,3,4,5]
var=pd.DataFrame(x)
print(var)
print(type(var))


# passing the dictionary
d={"a":[1,2,3,4,5],"b":[6,7,8,9,10],"c":[11,12,13,14,15],1:[1,2,3,4,5]}
var=pd.DataFrame(d)
print(var)

# restricting the columns
var=pd.DataFrame(d,columns=["a",1])
print(var)

# changig the indexes
var=pd.DataFrame(d,index=['A','B','C','D','E'])
print(var)

# accessing the data
print(var["a"][2])


# creating dataframe using the nested lists
list1=[[1,2,3,4,5],[5,6,7,8,9],["a","b","c","d","e"]]
var2=pd.DataFrame(list1)
print(var2)

# dataframes using the adding series
dict={"A":pd.Series([1,2,3,4,]) ,"B":pd.Series(["a","b","c","d"])}
var=pd.DataFrame(dict)
print(var)
```

```python
# arithmetic operators
import pandas as pd

dic = {"A": [1, 2, 3, 4], "B": [5, 6, 7, 8]}
var = pd.DataFrame(dic)
print(var)

var["C"] = var["A"] + var["B"]  # This should work correctly
print(var)
var["C"] = var["A"] - var["B"]  # This should work correctly
print(var)
var["C"] = var["A"] * var["B"]  # This should work correctly
print(var)
var["C"] = var["A"] / var["B"]  # This should work correctly
print(var)

# using the comparison operator in the each column
var["A>=2"]=var["A"]>=2
print(var)
var["B<=6"]=var["B"]<=6
print(var)
```

```python
#ins=etion and deletion functions
var=pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
print(var)
var.insert(1,"insertion",["a","b","c","d"])
print(var)

# insertion using the slicing
var["sliced"]=var["A"][:2]
print(var)

# pop function for deletion
var2=pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8],"C":[9,10,11,12]})
var2.pop("B")
print(var2)

# using the del keywod
del var2["A"]
print(var2)
```

```python
# functions to work with the NAN values in the dataset
import pandas as pd
var = pd.read_csv("Heart Prediction Quantum Dataset.csv",nrows=20,usecols=[0,1,2,4,5])
print(var)
```

```python
# using dropna function to remove all the null values
var.dropna()
var.dropna(axis=0)
var.dropna(how="any")
var.dropna(subset=["Age"])
# to modify the original dataframe without making a new copy
var.dropna(inplace=True)
print(var)
var.dropna(thresh=1)
```

```python
# functions with using the fillna function
var = pd.read_csv("Heart Prediction Quantum Dataset.csv",nrows=20,usecols=[0,1,2,4,5])
print(var)
var.fillna("python")
# to fill according to the specified columns
var.fillna({"Age":"Default age","QuantumPatternFeature":"pta ni"})
# to fill the backward values in the null data
# var.fillna(method="ffill")
# var.fillna(120,inplace=True)
print(var)
var.fillna("python",limit=1)
```
