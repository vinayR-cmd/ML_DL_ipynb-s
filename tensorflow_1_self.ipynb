basic arithmetic operations

```python
# add
a=tf.constant([1,2,3])
b=tf.constant([4,5,6])

print(tf.add(a,b))

# element wise multiplication
print(tf.multiply(a,b))
# same for the subract and divide

# real matrix multiplication
a=tf.constant([[1,2],[3,4]])
b=tf.constant([[5,6],[7,8]])
print(tf.matmul(a,b))

# reduction operations
x=tf.constant([1,2,3,4])
print(tf.reduce_sum(x))
# same for mean, max., min and others as well

# axis wise
x=tf.constant([[1,2,3,4],
               [5,6,7,8]])
print(tf.reduce_sum(x, axis=1))  #row wise
print(tf.reduce_sum(x, axis=0))   #column wise

# power and exponential
x=tf.constant([[1,2,3,4],
               [5,6,7,8]],dtype=tf.float32)
print(tf.square(x))
print(tf.pow(x,3))
print(tf.exp(x))
print(tf.rank(x))
# same for the sqrt and logarithmic
```

```python
# activation fuctions with the tensors
x=tf.constant([1,2,3,4],dtype=tf.float32)
print(tf.nn.relu(x))
print(tf.nn.sigmoid(x))
print(tf.nn.softmax(x))

# comparisons operators always give the output a boolean value
y=tf.constant([2,3,4,5],dtype=tf.float32)
print(tf.equal(x,y))
print(tf.greater(x,y))

# /clipping and normalisation
print(tf.clip_by_value(x,0,2))
print(tf.nn.l2_normalize(x))
```

```python
# variable tensors
w=tf.Variable([1.0,2.0,3.0],dtype=tf.float32)
print(w)

w.assign([10.0,11.0,12.0])
print(w)

# add to variable
w.assign_add([1.0,2.0,3.0])
print(w)

# subtract
w.assign_sub([1.0,2.0,3.0])
print(w)
```

```python
# gradient

w=tf.Variable(4.0)
# bcos value hme chaNGE KRNI HAI

for i in range(5):
  with tf.GradientTape() as tape:  # it records operations dynamically and later uses it to compuet gradient
    y=(w-2)**2  #loss function  #forward pass

  grad=tape.gradient(y,w)   #this computes dy/dw   #backward pass
  w.assign_sub(0.1*grad)   # this is the learning step w=w-lr*grad

  print(w.numpy()	)  # convert tensor to normal python
```

```python
from __future__ import absolute_import , division , print_function, unicode_literals
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import clear_output
from six.moves import urllib

import tensorflow.compat.v2.feature_column as fc
import tensorflow as tf
```

```python
import seaborn as sns
df = sns.load_dataset('titanic')
df.head()
```

```python
X=df.drop('survived',axis=1)
y=df['survived']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y, random_state=42, test_size=0.2)
```

```python
dftrain = pd.concat([X_train, y_train], axis=1)
dftest = pd.concat([X_test, y_test], axis=1)
```

```python
X_train.head(3)
```

```python
dftrain.head()
```

```python
CATEGORICAL_COLUMNS = ['sex', 'sibsp', 'parch', 'class', 'embarked']
NUMERIC_COLUMNS = ['age', 'fare']

feature_columns = []
for feature_name in CATEGORICAL_COLUMNS:
  vocabulary = dftrain[feature_name].unique()
  feature_columns.append(fc.categorical_column_with_vocabulary_list(feature_name, vocabulary))

for feature_name in NUMERIC_COLUMNS:
  feature_columns.append(fc.numeric_column(feature_name, dtype=tf.float32))

print(feature_columns)
```

```python
dftrain['embarked'].unique()
```

input function


```python
def make_input_fn(data_df, label_df, num_epochs=10, shuffle=True, batch_size=32):
  def input_function():  # inner function, this will be returned
    ds = tf.data.Dataset.from_tensor_slices((dict(data_df), label_df))  # create tf.data.Dataset object with data and its label
    if shuffle:
      ds = ds.shuffle(1000) # randomize order of data
    ds = ds.batch(batch_size).repeat(num_epochs)  # split dataset into batches of 32 and repeat process for number of epochs
    return ds  # return a batch of the dataset
  return input_function  # return a function object for use

train_input_fn = make_input_fn(dftrain, y_train)  # here we will call the input_function that was returned to us to get a dataset object we can feed to the model
eval_input_fn = make_input_fn(X_test, y_test, num_epochs=1, shuffle=False)
```

```python
linear_est = tf.estimator.LinearClassifier(feature_columns=feature_columns)
# We create a linear estimtor by passing the feature columns we created earlier
```

```python
linear_est.train(train_input_fn)
result=linear_est.evaluate(eval_input_fn)

clear_output()
print(result['accuracy'])
result
```

```python
# for prediction
pred_dicts = list(linear_est.predict(eval_input_fn))
probs = pd.Series([pred['probabilities'][1] for pred in pred_dicts])

probs.plot(kind='hist', bins=20, title='predicted probabilities')
```
