```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (
    Conv2D, AveragePooling2D, Flatten,
    Dense, BatchNormalization, Dropout, Activation
)
```

```python
model=Sequential()

model.add(Conv2D(6, kernel_size=(5,5),padding='valid', input_shape=(32,32,1)))
model.add(BatchNormalization())
model.add(Activation('tanh'))
model.add(AveragePooling2D(padding='valid', pool_size=(2,2), strides=2))
model.add(Dropout(0.3))

model.add(Conv2D(6, kernel_size=(5,5),padding='valid'))
model.add(BatchNormalization())
model.add(Activation('tanh'))
model.add(AveragePooling2D(padding='valid', pool_size=(2,2), strides=2))
model.add(Dropout(0.3))

model.add(Flatten())


model.add(Dense(120))
model.add(BatchNormalization())
model.add(Activation('tanh'))
model.add(Dropout(0.5))

model.add(Dense(84))
model.add(BatchNormalization())
model.add(Activation('tanh'))
model.add(Dropout(0.5))

model.add(Dense(10, activation='softmax'))
```

```python
model.summary()
```
