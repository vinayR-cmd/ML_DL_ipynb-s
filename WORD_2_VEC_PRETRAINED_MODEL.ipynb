```python
import gensim.downloader as api

# Download the Google News Word2Vec model
# This might take some time depending on your internet connection
word_vectors = api.load("word2vec-google-news-300")
```

```python
# The model is already loaded into 'word_vectors' by gensim.downloader
# No need to load it again from a file path
model = word_vectors
```

```python
model.similarity('man','woman')
```
